```python
import torch
import torch.nn.functional as F

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

ckpt = torch.load("../models/mobilenetv2_final.pth", map_location=device)

class_names = ckpt["idx_to_class"]
num_classes = len(class_names)

model = get_mobilenet_model(num_classes, version="v2").to(device)
model.load_state_dict(ckpt["model_state"])
model.eval()



def predict_with_confidence(model, image_tensor, class_names, device):
    model.eval()
    image_tensor = image_tensor.unsqueeze(0).to(device)  # (1, 3, 224, 224)

    with torch.no_grad():
        logits = model(image_tensor)
        probs = F.softmax(logits, dim=1)

    confidence, pred_idx = torch.max(probs, dim=1)

    return {
        "predicted_class": class_names[pred_idx.item()],
        "confidence": confidence.item(),
        "all_probabilities": probs.squeeze().cpu().tolist()
    }
```
